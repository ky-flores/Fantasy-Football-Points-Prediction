# Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import requests
```

# Loading in Data

```python
positions = ['QB','WR','TE','RB']
dst = 'DST'
kickers = 'K'

weeks = []
for i in np.arange(1,19):
    weeks.append(str(i))

years = list(range(2016,2024))

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

data = pd.DataFrame()
for year in years:
    year_df = pd.DataFrame()
    for position in positions:
        position_df = pd.DataFrame()
        for week in weeks:
            week_df = pd.DataFrame()
            url = f'https://www.footballdb.com/fantasy-football/index.html?pos={position}&yr={year}&wk={week}&key=48ca46aa7d721af4d58dccc0c249a1c4'
            response = requests.get(url, headers=headers)

            dfs = pd.read_html(response.text)
            week_df = dfs[0]
            week_df['Week'] = week

            position_df = pd.concat([position_df, week_df], axis=0)

        position_df['Position'] = position
        year_df = pd.concat([year_df,position_df], axis=0)
    year_df['Year'] = year
    data = pd.concat([data,year_df], axis=0)
```

```python
kicker_data = pd.DataFrame()

for year in years:
    year_df = pd.DataFrame()  
    for week in weeks:
        url = f'https://www.footballdb.com/fantasy-football/index.html?pos=K&yr={year}&wk={week}&key=48ca46aa7d721af4d58dccc0c249a1c4'
        response = requests.get(url, headers=headers)

        dfs = pd.read_html(response.text)
        week_df = dfs[0]
        week_df['Week'] = week

        year_df = pd.concat([year_df, week_df], axis=0)  

    year_df['Position'] = 'K'
    year_df['Year'] = year
    kicker_data = pd.concat([kicker_data, year_df], axis=0)  

kicker_data
```

```python
dst_data = pd.DataFrame()

for year in years:
    year_df = pd.DataFrame()  
    for week in weeks:
        url = f'https://www.footballdb.com/fantasy-football/index.html?pos=DST&yr={year}&wk={week}&key=48ca46aa7d721af4d58dccc0c249a1c4'
        response = requests.get(url, headers=headers)

        dfs = pd.read_html(response.text)
        week_df = dfs[0]
        week_df['Week'] = week

        year_df = pd.concat([year_df, week_df], axis=0)  

    year_df['Position'] = 'DST'
    year_df['Year'] = year
    dst_data = pd.concat([dst_data, year_df], axis=0)  

dst_data
```

# Cleaning Data

```python
# Renaming columns in data dataframe
data2 = data.copy()
data2.columns = data2.columns.droplevel(0)
data2.columns = list(data2.columns[0:19]) + ['Week', 'Position','Year']
data2['Player'] = data2['Player'].str.split('.').str[0]
data2['Player'] = data2['Player'].str[:-1]
data2
```

```python
kicker_data['Player'] = kicker_data['Player'].str.split('.').str[0]
kicker_data['Player'] = kicker_data['Player'].str[:-1]
kicker_data
```

```python
dst_data2 = dst_data.copy()
dst_data2['Team'] = dst_data2['Team'].str[-3:]
dst_data2['Team'] = dst_data2['Team'].str.replace('s', '')

dst_data2['Team'].value_counts()
```

```python
data3 = data2.copy()
data3['target'] = data3.groupby('Player')['Pts*'].shift(-1)
data3[data3['Player'] == 'Mac Jones']
```

```python
dst_data2['Target'] = dst_data2.groupby('Team')['Pts*'].shift(-1)
dst_data2[dst_data2['Team'] == 'MIN']
```

```python
def assign_opponent(row):
    teams = row['Game'].split('@')
    team1_initials = teams[0]
    team2_initials = teams[1]

    if team1_initials not in row['Team']:
        return team1_initials
    else:
        return team2_initials

dst_data2['Opponent'] = dst_data2.apply(assign_opponent, axis=1)
dst_data2
```
